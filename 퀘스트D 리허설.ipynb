### 코랩 공유시 latency

1. 코랩 사본을 만든 사람이 타이핑한 내용은  
새로고침을 누르지 않아도 상대방 코랩 화면에 반영이 되긴 하는데  
빠르면 10초 늦으면 20초 뒤에 뜨고  
타이핑한 내용의 일부만 업데이트 된다.  
  
2. 상대방이 새로고침을 눌러야 빠르면 5초 늦으면 10초 정도 지난 후에 바로 상대방 코랩화면에 반영된다  
  
3. 개더에서 화면을 공유하면 실시간으로 변경내용이 서로 확인 가능하니까 문제는 없음
